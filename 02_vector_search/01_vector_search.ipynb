## Week 2: vector search

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer(
    "all-MiniLM-L6-v2"
)  # list of available models: https://sbert.net/docs/sentence_transformer/pretrained_models.html
```

```python
# encode the question
q1 = "can i join post start"
v1 = model.encode(q1)
```

```python
# encode the docs
d = "You don't need to register. You're accepted. You can also just start learning and submitting homework without registering."
dv = model.encode(d)
```

```python
# compare the query with the docs using dot product
v1.dot(dv)
```

```python
# now try an unrelated query:
q2 = "How to install Docker on Windows?"
v2 = model.encode(q2)

v2.dot(dv)
```

* vector search: similar texts get similar vectors, and a dot product tells us how similar.
* dot products of normalized vectors = cosine similarity
  * `cos(0) = 1` - vectors are the same
  * `cos(90) = 0` - vectors are independent
  * `cos(180) = -1` - vectors are exactly the opposite

### Load the data with helper script from week 1, and then create embeddings

```python
from ingest import load_faq_data

documents = load_faq_data()
```

```python
from tqdm.auto import tqdm

texts = []

for doc in documents:
    text = doc["question"] + " " + doc["answer"]
    texts.append(text)

batch_size = 50
vectors = []

for i in tqdm(range(0, len(texts), batch_size)):
    batch = texts[i : i + batch_size]
    batch_vectors = model.encode(batch)
    vectors.extend(batch_vectors)

len(vectors)
```
